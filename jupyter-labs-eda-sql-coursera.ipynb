## Tasks

Now write and execute SQL queries to solve the assignment tasks.

### Task 1

##### Display the names of the unique launch sites  in the space mission


```python
#  get the SPACEXTBL column names
Query = "select DISTINCT LAUNCH_SITE from SPACEXTBL"

results = pd.read_sql(Query, pconn)
results
```

### Task 2

##### Display 5 records where launch sites begin with the string 'CCA'


```python
Query = "select * from SPACEXTBL where LAUNCH_SITE LIKE 'CCA%' "

results = pd.read_sql(Query, pconn)
results
```

### Task 3

##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
Query = "select sum(PAYLOAD_MASS__KG_) AS Total_Payload_NASA from SPACEXTBL where CUSTOMER = 'NASA (CRS)'"

results = pd.read_sql(Query, pconn)
results
```

### Task 4

##### Display average payload mass carried by booster version F9 v1.1


```python
Query = "select avg(PAYLOAD_MASS__KG_) AS Average_Payload_F9v11 from SPACEXTBL where BOOSTER_VERSION LIKE 'F9 v1.1%'"

results = pd.read_sql(Query, pconn)
results
```

### Task 5

##### List the date when the first successful landing outcome in ground pad was acheived.

*Hint:Use min function*


```python
Query = "select DATE from SPACEXTBL Where LANDING__OUTCOME LIKE 'Success (ground pad)%' ORDER BY DATE ASC LIMIT 1"

results = pd.read_sql(Query, pconn)
results
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
Query = "select BOOSTER_VERSION,LANDING__OUTCOME, PAYLOAD_MASS__KG_ from SPACEXTBL \
Where PAYLOAD_MASS__KG_ BETWEEN 4000 and 6000  \
AND LANDING__OUTCOME LIKE 'Success (drone ship)%'"

results = pd.read_sql(Query, pconn)
results
```

### Task 7

##### List the total number of successful and failure mission outcomes


```python
Query = "select count(case when LANDING__OUTCOME LIKE '%Success%' then 1 end) as Success, \
                 count(case when LANDING__OUTCOME LIKE '%Failure%' then 1 end) as Failure \
    from SPACEXTBL"

results = pd.read_sql(Query, pconn)
results
```

### Task 8

##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
Query = "select BOOSTER_VERSION, PAYLOAD_MASS__KG_ from SPACEXTBL \
Where PAYLOAD_MASS__KG_ = (select max(PAYLOAD_MASS__KG_) from SPACEXTBL)"

results = pd.read_sql(Query, pconn)
results
```

### Task 9

##### List the failed landing_outcomes in drone ship, their booster versions, and launch site names for in year 2015


```python
Query = "select LANDING__OUTCOME, BOOSTER_VERSION, LAUNCH_SITE , Date from SPACEXTBL \
       Where LANDING__OUTCOME LIKE 'Failure (drone ship)%' and Year(Date) = 2015"

results = pd.read_sql(Query, pconn)
results
```

### Task 10

##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order


```python
Query = "select  DATE, \
                 count(case when LANDING__OUTCOME LIKE '%Success%' then 1 end) as Success, \
                 count(case when LANDING__OUTCOME LIKE '%Failure%' then 1 end) as Failure \
                 from SPACEXTBL \
                 WHERE DATE BETWEEN '2010-06-04' and '2017-03-20' \
                 group by DATE \
                 order by DATE ASC"

results = pd.read_sql(Query, pconn)
results
```
